```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1,2,3,4,5,6,7]
y = [34, 35, 46, 44, 40, 32, 35]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather Data')
plt.plot(x,y, color='green', linewidth=3, linestyle='dashed')
```

### Format string -- used to control visual aspects

```python
x = [1,2,3,4,5,6,7]
y = [34, 35, 46, 44, 40, 32, 35]
plt.plot(x, y, 'D--r') #r-clor, D-daimond, -- dashed line 
```

```python
plt.plot(x, y, color='#003045', marker='D', linestyle='dashed', markersize='10') #Keyword arguments
```

```python
plt.plot(x, y, color='#003045', alpha=0.5) #alpha --> transparency 1-opeque & 0-transparent
```

### axis, labels, Legend, Grid

```python
days = [1,2,3,4,5,6,7]
max_t = [50,51,52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(days, max_t, label='max')
plt.plot(days, min_t, label='min')
plt.plot(days, avg_t, label='avg')
plt.legend(loc='upper right', shadow='true') # default -- best
plt.grid()
```
