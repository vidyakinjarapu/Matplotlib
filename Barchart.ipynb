```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
Company = ['GOOGL', 'MSFT', 'AMZN', 'FB']
Revenue = [56, 65, 55, 60]
Profit = [12, 23, 14, 5]
```

```python
xpos = np.arange(len(Company)) #set x-position
xpos
```

```python
plt.xticks(xpos, Company) # replace xpos with the company name
plt.ylabel('Revenue(bn)')
plt.xlabel('Company Name')
plt.title('US Tech Stock')
plt.bar(xpos-0.2, Revenue, width=0.4, label='Revenue')
plt.bar(xpos+0.2, Profit, width=0.4, label='profit')
plt.legend()
```

### Horizontal bar chart

```python
plt.yticks(xpos, Company) # replace xpos with the company name
plt.xlabel('Revenue(bn)')
plt.ylabel('Company Name')
plt.title('US Tech Stock')
plt.barh(xpos-0.2, Revenue,height=0.4, label='Revenue')
plt.barh(xpos+0.2, Profit, height=0.4, label='profit')
plt.legend()
```
