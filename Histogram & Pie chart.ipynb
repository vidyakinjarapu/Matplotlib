# Histogram and pie chart

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_sugar = [133, 85, 90, 150, 149, 80, 93, 115, 135, 80, 77, 82, 129] # single dimention array and don't need names
```

```python
plt.hist(blood_sugar, bins=3, rwidth=0.90)  #bins = 10 & rwidth-- Relative width -- default--1
```

```python
plt.hist(blood_sugar, bins=[80, 100, 125, 150], rwidth=0.90, color='green', histtype='step') 
```

```python
blood_sugar_men = [133, 85, 90, 150, 149, 80, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [100, 95, 90, 140, 153, 90, 100, 130, 140, 85, 89, 77, 100]
plt.hist([blood_sugar_men, blood_sugar_women], bins=[80, 100, 125, 150], rwidth=0.90, color=['green', 'orange'],
         label=(['men', 'women']))
plt.xlabel('Sugar range')
plt.ylabel('No of patients')
plt.title("Health Data")

plt.legend()
```

```python
blood_sugar_men = [133, 85, 90, 150, 149, 80, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [100, 95, 90, 140, 153, 90, 100, 130, 140, 85, 89, 77, 100]
plt.hist([blood_sugar_men, blood_sugar_women], bins=[80, 100, 125, 150], rwidth=0.90, color=['green', 'orange'],
         label=(['men', 'women']), orientation = 'horizontal') #orientation--> to draw plot horizontally
plt.xlabel('Sugar range')
plt.ylabel('No of patients')
plt.title("Health Data")

plt.legend()
```

# Pie Chart

```python
exp_val = [1400, 600, 400, 410, 250]
exp_labels = ['Home rent', 'Food', 'Internet bill', 'car', 'Other utilities' ]
```

```python
plt.axis("equal") # to get perfect circle
plt.pie(exp_val, labels=exp_labels)
```

```python
plt.axis("equal")
plt.pie(exp_val, labels=exp_labels, radius=1.5, autopct='%0.2f%%', shadow=True, explode=[0,0.3,0,0,0], startangle=45)
#radious to increse the size of the pie chart & autopct = to show percentages on plot--0.2 is 2 decimal points
plt.show() # to git rid of all unnecessary lines    
```
